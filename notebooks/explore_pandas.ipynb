```python
import pandas as pd
```

```python
df = pd.read_csv('../results/image_classes.csv')
df
```

```python
# Pasted from your google sheet. Use window key + shift + down arrow to create multiple cursors, then ctrl + side arrows to move across items
# Change this based on your data
reviewed_rows ={
10: 	"non_target",
121: 	"broadleaf",
231: 	"broadleaf",
}

# Loop through csv and update the values
for i, row in df.iterrows():
    if i in reviewed_rows:
        print("Before:", i, df.at[i, 'non_target_weed_class'])
        print("Image:", df.at[i, 'image_id'])
        df.at[i, 'non_target_weed_class'] = reviewed_rows[i]
        print("After:", i, df.at[i, 'non_target_weed_class'])
```
