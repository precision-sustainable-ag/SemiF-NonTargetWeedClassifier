```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
master_csv = Path('../results/image_classes.csv')
df = pd.read_csv(master_csv,low_memory=False)
```

```python
# Distribution of classes in the dataset
df.groupby('non_target_weed_class').size().plot(kind='bar')
```

```python
# Distribution of areas by classifier class
area_class_order = ["very_large", "large", "medium_large", "medium", "small", "very_small"]
plt.figure(figsize=(12, 6))
sns.countplot(
    data=df,
    x="non_target_weed_class",
    hue="area_category",
    hue_order=area_class_order,
    order=df["non_target_weed_class"].value_counts().index,
    palette="viridis"
)

plt.title("Distribution of bbox_area_cm2 by Non target weed class and Area Class", fontsize=16)
plt.xlabel("Non target weed class", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.legend(title="Area Class", title_fontsize=12, fontsize=10)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Distribution of areas by common name
area_class_order = ["very_large", "large", "medium_large", "medium", "small", "very_small"]
plt.figure(figsize=(12, 6))
sns.countplot(
    data=df,
    x="category_common_name",
    hue="area_category",
    hue_order=area_class_order,
    order=df["category_common_name"].value_counts().index,
    palette="viridis"
)

plt.title("Distribution of bbox_area_cm2 by Common name and Area Class", fontsize=16)
plt.xlabel("Common name", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.legend(title="Area Class", title_fontsize=12, fontsize=10)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```
